## Extract information from the given web site

You will extract the data from the below web site: <br>


```python
#this url contains the data you need to scrape
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/Programming_Languages.html"
```

The data you need to scrape is the **name of the programming language** and **average annual salary**.<br> It is a good idea to open the url in your web broswer and study the contents of the web page before you start to scrape.


Import the required libraries


```python
# Your code here
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

Download the webpage at the url


```python
#your code goes here
data=requests.get(url).text
```

Create a soup object


```python
#your code goes here
soup=BeautifulSoup(data,'html5lib')
```

Scrape the `Language name` and `annual average salary`.


```python
#your code goes here
la=[]
table=soup.find('table')
for row in table.find_all('tr'):
    cols=row.find_all('td')
    Lan_name=cols[1].getText()
    sal=cols[3].getText()
    la.append([Lan_name,sal])
```

Save the scrapped data into a file named *popular-languages.csv*


```python
# your code goes here
df=pd.DataFrame(la,columns=["Language Name","Average Annual Salary"])
df.drop(0,axis=0,inplace=True)
df.set_index("Language Name",inplace=True)
df['Average Annual Salary'].replace('[\$,]', '', regex=True, inplace=True)
df['Average Annual Salary'] = df['Average Annual Salary'].astype(float)


df.to_csv("popular-languages.csv")

from project_lib import Project

project = Project(None, '1c6a2917-9e15-4c3d-892c-f6929fa9d8be', 'p-64eaf951bb919826fe0677d4897b53202d6cdb9c')
import io

filename = 'popular-languages.csv'
with open(filename, 'rb') as z:
        data = io.BytesIO(z.read())
        project.save_data(
            filename, data, set_project_asset=True, overwrite=True
        )
```
